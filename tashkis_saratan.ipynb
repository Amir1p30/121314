```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

data=pd.read_csv('cancer.csv')
data = data.drop(['Unnamed: 32', 'id'], axis = 1)

index=0
for i in data.diagnosis:
    if i=='M':
        data.diagnosis[index]=1
    else:
        data.diagnosis[index]=0
    index=index+1

y = data.diagnosis
y = y.astype('int')
x_data = data.drop(['diagnosis'], axis = 1)

x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15)

models=(LogisticRegression(),DecisionTreeClassifier(),GaussianNB(),KNeighborsClassifier())


for Algorithm in models:
    model=Algorithm
    model.fit(x_train,y_train)
    out=model.score(x_test,y_test)
    print(f'the algorithm {Algorithm} and the result is {out}')


# model = LogisticRegression(max_iter=100000)
# model.fit(x_train, y_train)
# out_robot=model.predict(x_test)
```
